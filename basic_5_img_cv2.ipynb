```
import cv2
import matplotlib.pyplot as plt
import seaborn as sns
```

```
before_img = cv2.imread("/content/IMG_20250816_123542.jpg",cv2.IMREAD_GRAYSCALE)
```

```
before_img
```

```
before_img.shape
```

```
sns.heatmap(before_img,cmap="gray")
```

```
sns.heatmap(before_img, cmap="coolwarm")
```

```
plt.matshow(before_img, cmap="gray")
```

```
plt.imshow(before_img, cmap="gray")
```

```
before_img.shape
```

```
img_resized = cv2.resize(before_img,(200,200))
```

```
img_resized
```

```
img_resized.shape
```

```
sns.heatmap(img_resized, cmap="gray")
```

```
img_resized = cv2.bitwise_not(img_resized)
img_resized
#bitwise_not dile ultte zay 0 thakle 255 or 255 tahkle 0 erokom hoy
```

```
plt.imshow(img_resized, cmap="gray")
```
